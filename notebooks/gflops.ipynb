```python
import torch
import torch.nn as nn

class Encoder(nn.Module):
    def __init__(self, latent_dim=64, noise_std=0.1):
        super().__init__()
        self.encoder = nn.Linear(28*28, latent_dim)
        self.noise_std = noise_std
        
class FC1(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(64, 64)
        
class FC2(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc2 = nn.Linear(64, 10)

def linear_flops(in_features, out_features, batch_size=1, relu=False):
    flops = 2 * in_features * out_features
    if relu:
        flops += out_features
    return flops * batch_size
def flops_per_module(batch_size=32):
    flops_encoder = linear_flops(28*28, 64, batch_size) + 2*64*batch_size  # Gaussian noise
    flops_fc1 = linear_flops(64, 64, batch_size, relu=True)
    flops_fc2 = linear_flops(64, 10, batch_size)
    
    print(f"Encoder FLOPs: {flops_encoder}")
    print(f"FC1 FLOPs: {flops_fc1}")
    print(f"FC2 FLOPs: {flops_fc2}")
    print(f"Total FLOPs: {flops_encoder + flops_fc1 + flops_fc2}")

flops_per_module(batch_size=32)
```
